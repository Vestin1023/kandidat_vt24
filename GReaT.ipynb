```python
from be_great import GReaT

import pandas as pd
data = pd.read_csv('CompetenceContractorFullJoin.csv')

model = GReaT(llm='distilgpt2', batch_size=64, epochs=200, save_steps=400000)
model.fit(data)
synthetic_data = model.sample(n_samples=100)
synthetic_data.head(3)
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = GReaT(llm='distilgpt2', batch_size=64, epochs=200, save_steps=400000, device=device)

model.fit(data)

synthetic_data = model.sample(n_samples=100, device=device)
```

```python
synthetic_data.head(3)
synthetic_data.to_csv('synthetic_data_GReaT.csv', index=False)
```
